Importing the Essential Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

Analysing the Dataset and Preprocessing the data

```python
dataset = pd.read_csv("CarPrice_Assignment.csv")
print("Shape of the dataset:",dataset.shape)
```

```python
dataset['fueltype'].value_counts()
```

```python
dataset.replace({'fueltype':{'gas':1,'diesel':0}}, inplace=True)
```

```python
dataset['aspiration'].value_counts()
```

```python
dataset.replace({'aspiration':{'std':0,'turbo':1}}, inplace=True)
```

```python
dataset['doornumber'].value_counts()
```

```python
dataset.replace({'doornumber':{'four':4,'two':2}}, inplace=True)
```

```python
dataset['carbody'].value_counts()
```

```python
dataset.replace({'carbody':{'sedan':0,'hatchback':1,'wagon':2,'hardtop':3,'convertible':4}}, inplace=True)
```

```python
dataset['drivewheel'].value_counts()
```

```python
dataset.replace({'drivewheel':{'fwd':0,'rwd':1,'4wd':2}}, inplace=True)
```

```python
dataset['enginelocation'].value_counts()
```

```python
dataset.replace({'enginelocation':{'front':0,'rear':1}}, inplace=True)
```

```python
dataset['enginetype'].value_counts()
```

```python
dataset.replace({'enginetype':{'ohc':0,'ohcf':1,'ohcv':2,'dohc':3,'l':4,'rotor':5,'dohcv':6}}, inplace=True)
```

```python
dataset['cylindernumber'].value_counts()
```

```python
dataset.replace({'cylindernumber':{'four':4,'six':6,'five':5,'eight':8,'two':2,'three':3,'twelve':12}}, inplace=True)
```

```python
dataset['fuelsystem'].value_counts()
```

```python
dataset.replace({'fuelsystem':{'mpfi':0,'2bbl':1,'idi':2,'1bbl':3,'spdi':4,'4bbl':5,'mfi':6,'spfi':7}}, inplace=True)
```

Creating the features and dependent variable

```python
X = dataset.iloc[:,3:-1].values
y = dataset.iloc[:,-1].values
y = y.reshape(len(y),1)
```

```python
X
```

```python
y
```

Feature Scaling the Dependent Variables and Features

```python
from sklearn.preprocessing import StandardScaler
scx = StandardScaler()
scy = StandardScaler()

X = scx.fit_transform(X)
y = scy.fit_transform(y)
```

Splitting the dependent variable and features into training and testing dataset

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2,random_state=0)
```

```python
y_train
```

```python
y_test
```

Implementing SVR Machine Learning Model

```python
from sklearn.svm import SVR
model = SVR(kernel='rbf')
model.fit(X_train,y_train.ravel())
```

Predicting the Test Results

```python
y_pred = scy.inverse_transform([model.predict(X_test)])
print(y_pred)
```
